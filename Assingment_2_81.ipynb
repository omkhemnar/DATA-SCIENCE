```python
import pandas as pd
import matplotlib.pyplot as plt

data= {
    'Sub-Name': ['Marathi', 'Hindi', 'English', 'Math', 'Science', ],
    'Branch': ['IT', 'IT', 'MECH', 'MECH', 'CIVIL'],
    'Student': ['', 'Bob', 'Charlie', 'David', 'Eva'],
    'Div': ['A', 'A', 'C', 'C', 'E'],
    'Sub-Code': ['M101', 'H102', 'E103', 'M104', 'S105'],
    'Sub-Teacher': ['Mr. A', 'Ms. B', 'Dr. C', 'Prof. D', 'Dr. E'],
    'Marks': [85, 90, 78, 88, 92],}

df=pd.DataFrame(data)
df.head()
```

```python
# Bar chart for Division distribution
df['Div'].value_counts().plot(kind='bar', color=['lightgreen', 'lightblue', 'orange', 'violet', 'pink'])
plt.title("Division Distribution")
plt.xlabel("Division")
plt.ylabel("Count")
plt.show()
```

```python
# Bar chart for Branch distribution
df['Branch'].value_counts().plot(kind='bar', color=['skyblue', 'lightgreen', 'orange'])
plt.title("Branch Distribution")
plt.xlabel("Branch")
plt.ylabel("Count")
plt.show()
```

```python
# --- QUANTITATIVE ANALYSIS ---
print("\nQuantitative Analysis:")
print(df['Marks'].describe())  # Summary statistics

# Histogram for Marks
df['Marks'].plot(kind='hist', bins=5, color='skyblue', edgecolor='black')
plt.title("Marks Distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()

# If you want a scatter plot, you need another quantitative column.
# Example: Marks vs Index (just for demonstration)
plt.scatter(df.index, df['Marks'], color='green')
plt.title("Index vs Marks")
plt.xlabel("Index")
plt.ylabel("Marks")
plt.show()
```
